```python
import pandas as pd
import numpy as np

# Load original raw data
df = pd.read_csv("C:\\Users\\pokal\\OneDrive\\Dokumen\\Hospital-Readmission-Project\\data\\diabetic_data.csv")
print(f"Original shape: {df.shape}")

# Step 1 - Drop irrelevant columns
df = df.drop(columns=['encounter_id', 'patient_nbr', 'examide', 'citoglipton'])

# Step 2 - Replace ? with NaN
df = df.replace('?', np.nan)

# Step 3 - Drop high missing columns only
df = df.drop(columns=['weight', 'payer_code', 'medical_specialty'])

# Step 4 - Fill missing values instead of dropping rows
df['race'] = df['race'].fillna('Unknown')
df['diag_1'] = df['diag_1'].fillna('Unknown')
df['diag_2'] = df['diag_2'].fillna('Unknown')
df['diag_3'] = df['diag_3'].fillna('Unknown')

# Step 5 - Fix target variable
df['readmitted'] = df['readmitted'].apply(lambda x: 1 if x == '<30' else 0)

# Step 6 - Fix age column
age_map = {
    '[0-10)': 5, '[10-20)': 15, '[20-30)': 25,
    '[30-40)': 35, '[40-50)': 45, '[50-60)': 55,
    '[60-70)': 65, '[70-80)': 75, '[80-90)': 85, '[90-100)': 95
}
df['age'] = df['age'].map(age_map)

# Step 7 - Remove duplicates
df = df.drop_duplicates(subset='patient_nbr' if 'patient_nbr' in df.columns else None)

print(f"Final shape: {df.shape}")
print(f"Readmission distribution:\n{df['readmitted'].value_counts()}")

# Save
df.to_csv("C:\\Users\\pokal\\OneDrive\\Dokumen\\Hospital-Readmission-Project\\exports\\cleaned_data.csv", index=False)
print("Cleaned data saved ✅")
```
